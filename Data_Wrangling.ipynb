```python
count_dict = {}

for column_name in df:
    count = 0
    for value in df[column_name]:
        if pd.isnull(value):
            count += 1
    count_dict[column_name] = count

count_dict

            
        
```

```python
for i in df:
    for j in df[i]:
        if pd.isnull(j):
           df[i].replace(j,'np.Nan')
```

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")  
```

```python
mean= df['Weight_kg'].mean()
mean
```

```python
df['Weight_kg'].replace((np.nan,mean),inplace=True)
```

```python
avg_weight=df['Weight_kg'].astype('float').mean(axis=0)
df["Weight_kg"].replace(np.nan, avg_weight, inplace=True)
```

```python
df['Screen_Size_cm'].value_counts()
df['Screen_Size_cm'].replace(np.nan,40.0,inplace=True)
```

```python
df['Weight_kg']=df['Weight_kg'].astype('float')
```

```python
df['Screen_Size_cm']=df['Screen_Size_cm'].astype('float')
```

```python
df['Screen_Size_cm']=df['Screen_Size_cm']/2.54
```

```python
df.rename(columns={'Screen_Size_cm':'Screen_Size_inches'},inplace=True)
```

```python
df['Weight_kg']=df['Weight_kg']*2.205
```

```python
df['CPU_frequency']=df['CPU_frequency']/df['CPU_frequency'].max()
```

```python
bins=np.linspace(min(df['Price']),max(df['Price']),4)
bins
```

```python
cols=['min','medium','high']
```

```python
df['Price']=pd.cut(df['Price'],bins,labels=cols, include_lowest=True)
```

```python
plt.bar(cols, df["Price"].value_counts())
plt.xlabel("Price")
plt.ylabel("count")
plt.title("Price bins")
```
